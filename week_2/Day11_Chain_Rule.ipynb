# 链式法则与全微分学习笔记

## 一、链式法则
1.  `sp.Symbol('x')`：创建的是一个抽象的数学符号，就像我们在草稿纸上写的未知数 $x$，可以用来做代数运算、求导、解方程等符号操作。
2.  **多层一元复合函数链式法则**
    设 $y = f(u),\ u = \varphi(v),\ v = \psi(x)$，则：
    $$
    \frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dv} \cdot \frac{dv}{dx}
    $$
3.  **必须用 `subs()` 替换符号变量**
    链式法则得到的结果里还保留着中间变量（如 `m` 和 `n`），必须用 `.subs({m:m_t, n:n_t})` 把它们替换成具体的 $t$ 表达式，才能和直接求导的结果比较。
4.  **推荐用 `sp.expand()` 展开验证**
    即使不做比较，用 `sp.expand()` 把两个结果都展开成标准多项式，也能更直观地看到它们是同一个表达式。
5.  **用 `sp.simplify()` 做差验证**
    `sp.simplify(df_t_sub - df_t_direct) == 0` 是最可靠的验证方式，只要差为 0，就说明结果完全一致。

## 二、全微分 必背公式
1.  **二元函数全微分**
    设 $z = f(x,y)$：
    $$
    dz = \frac{\partial z}{\partial x}dx + \frac{\partial z}{\partial y}dy
    $$
2.  **三元函数全微分**
    设 $u = f(x,y,z)$：
    $$
    du = \frac{\partial u}{\partial x}dx + \frac{\partial u}{\partial y}dy + \frac{\partial u}{\partial z}dz
    $$
3.  **可微充分条件**
    一阶偏导数连续 $\implies$ 函数可微。
4.  **全增量近似公式（近似计算）**
    $$
    \Delta z \approx dz = f'_x(x_0,y_0)\Delta x + f'_y(x_0,y_0)\Delta y
    $$
5.  **全微分形式不变性**
    设 $z = f(u,v)$：
    $$
    dz = \frac{\partial z}{\partial u}du + \frac{\partial z}{\partial v}dv
    $$
    $u、v$ 不管是自变量还是中间变量，公式不变。
6.  **四则运算微分**
    $$
    \begin{align*}
    d(u \pm v) &= du \pm dv \\
    d(uv) &= vdu + udv \\
    d\left(\frac{u}{v}\right) &= \frac{vdu - udv}{v^2}
    \end{align*}
    $$

```python
import sympy as sp
import math
x,y=sp.symbols('x,y')
z=(y**3)*(2**x+3*y)
sp.init_printing()
dx=sp.diff(z,x)
dy=sp.diff(z,y)
print("\n全微分dz=",dx*sp.Symbol('dx')+dy*sp.Symbol('dy'))
#链式法则
n,m,t=sp.symbols('n,m,t')
m_t=2*t**3
n_t=t*(t+3)
f=m**2+n**2
#1.链式法则直接求导
df_t_direct=sp.diff(f.subs({m:m_t,n:n_t}),t)
#2.链式法则分布求导
dm=sp.diff(f,m)
dn=sp.diff(f,n)
dm_t=sp.diff(m_t,t)
dn_t=sp.diff(n_t,t)
df_t=dm*dm_t+dn*dn_t
is_equal=(df_t==df_t_direct)
#相等的关键：把链式法则结果里的m，n全部替换成具体表达式
df_t_sub=df_t.subs({m:m_t,n:n_t})
sp.init_printing()
print("链式法则求导t：",sp.expand(df_t_sub))
print("直接求导：",sp.expand(df_t_direct))
print("两者是否相等：",sp.simplify(df_t_sub-df_t_direct)==0)
```
