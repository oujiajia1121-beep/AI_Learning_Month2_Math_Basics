偏导数

```python
import sympy as sp
x,y=sp.symbols('x,y')
f=x**2+y**2
diffed_x=sp.diff(f,x)
sp.init_printing()
print(diffed_x)
diffed_y=sp.diff(f,y)
print(diffed_y)
term1=x**2
term2=y**2
diffed_term1_x=sp.diff(term1,x)
diffed_term2_x=sp.diff(term2,x)
df_x=diffed_term1_x+diffed_term2_x
is_equal=(diffed_x==df_x)
print("两种方法是否相等：",is_equal)
```
