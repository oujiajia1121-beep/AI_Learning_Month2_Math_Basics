1. SymPy = Python专业数字符号计算库：专门用来计算公式、解方程、求导、积分、化简代数式，不是算小数，是手写数学式子一样运算
2. sp.symbols('变量') ：定义x，y为数学变量（不是普通数字）
3. sp.init_printing() : 美化输出，显示标准数学格式(不需要传入变量，他只是开启美化输出模式，不是用来打应变量的,一个代码里只用写一次，
后边的都会自动简化)
4. sp.simplify(式子) ：自动约分、合并、简化复杂代数式（分式、繁琐多项式快速整理）
5. 求导 diff : sp.diff(式子)（作用：默认为求一阶导数） 二阶求导：sp.diff(式子,x,2) （所以高阶求导就改一下数字就可以了）

```python
import sympy as sp
x=sp.symbols('x')
y=(x**2-1)/(x-1)
simplified_y=sp.simplify(y)
sp.init_printing()
print(simplified_y)
z=x**3+2*x+8
diffed_z=sp.diff(z,x)
print(diffed_z)

```
