使用NumPy计算一个2*2矩阵乘以一个2*1向量
1. np.dot() 矩阵乘法，标准矩阵惩罚的核心维度要求就是前一个矩阵的列数等于后一个矩阵的行数

```python
import numpy as np
arr1=np.array([[1,2],[3,4]])
arr2=np.array([2,2])
print(np.dot(arr1,arr2))
print(arr1.shape)
print(arr2.shape)
```

用 np.linalg.inv 和 np.linalg.solve 求解一个二元一次方程组
1. np.linalg.inv(arr) ：矩阵求逆函数，函数解可以看做 Ax=y,这个函数求的就是A逆，要求就是输入必须是矩阵，且矩阵可逆
2. a@b 在这里表示a*b
3. np.linalg.solve(a,b) : 线性方程求解函数

```python
import numpy as np
a=np.array([[2,3],
            [1,-2]])
b=np.array([13,4])
a_inv=np.linalg.inv(a)
x=a_inv@b
y=np.linalg.solve(a,b)
print("解为：",x,"\n",y)
```
