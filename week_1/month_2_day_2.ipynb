```python
import numpy as np
a=np.array([2,3,8])
b=np.array([4,6,16])
f=np.array([2,3,8])
#计算点积
c=np.dot(a,b)
h=np.dot(a,f)
#手动验算
d=2*4+3*3+8*9
is_equal=(c==d)
print("点积结果：",c)
print("手动验算结果：",d)
print("验算结果：",is_equal)
print(h)
```

一、 用NumPy计算两个向量的点积
1. 点积的计算公式 a·b=a1b1+......+anbn
2. np.dot(a,b) 计算：a·b
3. 当只有单个数之间判断相等，可以直接 (a==b) 会返回 True or False
4. print(a@b) 和上面计算点积的效果一样
