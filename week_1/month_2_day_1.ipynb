```python
import numpy as np
a=np.array([1,2,3])
b=np.array([4,5,6])
c=a+b
e=b-a
print("向量e：",e)
print("向量a：",a)
print("向量b：",b)
print("a+b=",c)
#手动检验
d=np.array([1+4,2+5,3+6])
f=np.array([4-1,5-2,6-3])
print("手动验算结果d：",d)
print("手动验算结果f：",f)
#判断是否相等
is_equal=np.all(c==d)
is_equal_2=np.all(e==f)
print("加法验证是否成立：",is_equal,is_equal_2)
```

一、 使用NumPy库创建向量并验证加法和减法
1. np.array([x,y,z])生成一对数组=数学中的向量,array 是数组的意思
2. Numpy向量加法为对应位置元素逐个相加
3. np.all()用于整体比对两个向量是否完全一致
