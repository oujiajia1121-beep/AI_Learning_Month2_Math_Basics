## 概率的基本概念
 1. np.random.choice(序列,数量,概率) ： 按指定概率随机选取

 2. 频率=正面次数/总实验次数

 3. 满足大数定理：实验次数越大，频率无线逼近真实概率

 4. np.sum(result==1) : 当满足=True（数值=1），不满足=False（数值=0）

```python
import numpy as np
# 1. 设置参数：1=正面，0=反面，概率各0.5
times=100
coins=[1,0]
p=[0.5,0.5]
# 2. np.random.choice 随机抽样模拟抛硬币
result=np.random.choice(coins,size=times,p=p)
# 3. 统计正面次数 & 频率
front_count=np.sum(result==1)
front_freq=front_count/times
# 4. 输出结果
print(f"抛硬币总次数：{times}")
print(f"正面朝上次数：{front_count}")
print(f"正面朝上频率：{front_freq:.4f}")
print("理论概率：0.5，次数越多频率趋近0.5")
# 扩大次数
times_big=10000
res_big=np.random.choice([1,0],size=times_big,p=[0.5,0.5])
freq_big=np.sum(res_big==1)/times_big
print(f"\n10000次正面频率：{freq_big:.4f}")
```

## 自己动手拓展

1. plt.figure(figsize=(10,6)) ： 创建一个指定大小的画布，后续的绘画都画在这个画布上 figsize=(10,6):表示画布的宽度是10英寸，高度是6英寸，宽高比为10:6（也就是5:3）

2. plt.plot(x,[theoretical_prob]*len(x),'b-') ： [theoretical_prob]*len(x) 把单个的0.5复制成和x长度一样的列表，这样才能和x轴数据配对，画出一条贯穿整个图表的水平线

3. plt.scatter(): 画散点图，展示每次实验的实际评率；alpha=0.5 让点半透明，重叠的地方更清楚

```python
import numpy as np
import matplotlib.pyplot as plt
def simulate_coin_flip(coins):
    times=[]
    freqs=[]
    for n in range(10,10000):
        p=[0.5,0.5]
        result=np.random.choice(coins,size=n,p=p)
        front_count=np.sum(result==1)
        front_freq=front_count/n
        freqs.append(front_freq)
        times.append(n)
    return times,freqs
coins=[1,0]
theoretical_prob=0.5
plt.figure(figsize=(10,6))
x,y=simulate_coin_flip(coins)
plt.plot(x,[theoretical_prob]*len(x),'b-')
plt.scatter(x,y,c='r',s=10,alpha=0.5)
plt.xlabel('times')
plt.ylabel('frequence')
plt.show()
```
