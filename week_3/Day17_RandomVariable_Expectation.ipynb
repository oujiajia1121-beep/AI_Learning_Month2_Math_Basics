# 随机变量与期望

## 随机变量（Random Variable）：

1. 离散随机变量（Discrete Random Variable）：

·取数是有限个或可数无限个
                                             
· 用 概率质量函数 （PMF）描述

2. 连续随机变量（Continuous Random Variable）
   
   · 取值是一个连续区间内的任意实数

   · 用概率密度函数（PDF）描述，曲线下面积表示概率

## 期望值（Expected Value）：

   · 期望值是随机变量所有可能值的加权平均数，权重是其对应的概率

   ·公式 ：每个可能值乘以它出现的概率，再求和

   · 如果重复试验无数次，样本均值会趋近于期望值，这正是 “大数定律” 的核心

## 期望的线性性质（重要性质）
 
   · 期望值有线性性质，对任意随机变量 X,Y 和常数 a，b 有：  E[aX + bY] = a·E[X] + b·E[Y] 

## 方差（Variance）

   · 方差衡量随机变量取值相对于其期望值的离散程度/波动大小 ： Var[X] = E[(X-mu)**2] = E[X**2]  - (E[X])**2  (E[X**2] 表示把随机变量 X 取的每一个值都平方，再求这些平方值的期望)

    其中 mu=E[X] 为期望值。方差越大，说明数据波动越大

3. np.var(): 可以直接计算方差

4. 函数参数的优先级顺序（必须遵守）：Python规定，调用函数时，参数必须遵守这个顺序写：

   位置参数 → 可变位置参数 → 关键字参数（默认参数） → 可变关键字参数

5. plt.axhline() / plt.axvline() （水平线 / 垂直线）   

```python
import numpy as np
import matplotlib.pyplot as plt

# 理论值
X = np.array([1,2,3,4,5,6])
P = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
E_x_theory = np.sum(X * P)                    # 3.5
Var_x_theory = np.sum((X - E_x_theory)**2 * P) # 约2.9167

print(f"理论期望: {E_x_theory}")
print(f"理论方差: {Var_x_theory:.4f}")

# 模拟
n_values = range(10, 10001, 10)   # 试验次数列表
mean_list = []
var_list = []

for n in n_values:
    result = np.random.choice(X, size=n, p=P)
    mean_list.append(np.mean(result))
    var_list.append(np.var(result))   # 直接用 np.var 计算方差（总体方差）

# 绘图
plt.figure(figsize=(10,5))
plt.plot(n_values, mean_list, 'b-', label='模拟均值')
plt.axhline(y=E_x_theory, color='b', linestyle='--', label='理论期望 (3.5)')
plt.plot(n_values, var_list, 'r-', label='模拟方差')
plt.axhline(y=Var_x_theory, color='r', linestyle='--', label=f'理论方差 ({Var_x_theory:.4f})')
plt.xlabel('试验次数')
plt.ylabel('值')
plt.title('掷骰子：累积均值和方差收敛')
plt.legend()
plt.show()

print(f"最终（n=10000）模拟均值: {mean_list[-1]:.4f}, 理论: {E_x_theory}")
print(f"最终模拟方差: {var_list[-1]:.4f}, 理论: {Var_x_theory:.4f}")
```

```python
## 掷骰子的期望和方差（我自己做的）

import numpy as np
import matplotlib.pyplot as plt

X=np.array([1,2,3,4,5,6])
P=np.array([1/6,1/6,1/6,1/6,1/6,1/6])
E_x=np.mean(X)
Var_x=np.mean(X**2) - np.mean(X)**2
print(f"理论期望：{E_x}")
print(f"理论方差：{Var_x}")

# 模拟验证

mean_list=[]
var_list=[]
times=range(10,10001,10)
for n in times:
    result=np.random.choice(X,size=n,p=P)
    mean_list.append(np.mean(result))
    var_list.append(np.var(result))
plt.figure(figsize=(10,6))
plt.plot(times,mean_list,'b-')
plt.axhline(y=E_x,color='b',linestyle='--')
plt.plot(times,var_list,'r-')
plt.axhline(y=Var_x,color='r',linestyle='--')
plt.show()

```
